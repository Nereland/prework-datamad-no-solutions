```python
# Import the choice function of the random module 
import random
# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
game = ["rock", "paper", "scissors"]
# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
maxgames = 4
# Assign a variable to the number of games a player must win to win. Preferably the value will be based on the number of maximum games
winstowin = int(maxgames/2 + 1) 

# Define a function that randomly returns one of the 3 options. This will correspond to the play of the machine. Totally random.

def machineChoice():
    return random.choice(game)

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def userChoice():
    user_choice = ""
    while user_choice not in game:
       user_choice = input("Choose rock, paper or scissors: ")
    return user_choice


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
playerwins = 0
computerwins = 0
tie = 0

def onecombat(): 
    player = userChoice()
    computer = machineChoice()
    print("computer chose: ", computer)

    global playerwins
    global computerwins
    global tie

    if player == computer:
        tie +=1
        return 0 
    elif player == "rock":
        if computer == "scissors":
            playerwins+= 1
            return 2
        else: 
            computerwins+= 1
            return 1
    elif player == "paper":
        if computer == "rock":
            playerwins+= 1
            return 2
        else:
            computerwins+= 1
            return 1
    elif player == "scissors":
        if computer == "paper":
            playerwins+= 1
            return 2
        else:
            computerwins+= 1
            return 1


# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
# Create two variables that accumulate the wins of each participant
def battle():
    rounds = 0
    while rounds < maxgames:
        rounds += 1
        result = onecombat()
        print("round ", rounds, "out of", maxgames)
        print("Last combat result (1 = machine won, 2 = player won, 0 = tie):", result)
        print("playerpoints: ", playerwins)
        print("computerpoints: ", computerwins)
        print("ties: ", tie)
        if  playerwins == winstowin:
            return "HUMAN PLAYER WON THE BATTLE!"
        elif computerwins == winstowin:
            return "THE COMPUTER WON THE BATTLE!"            
        elif rounds == maxgames:
            if playerwins > computerwins:
                return "HUMAN PLAYER WON THE BATTLE!"
            elif computerwins > playerwins:
                return "THE COMPUTER WON THE BATTLE!" 
            else:
                return "DRAW!"


#print(pointstowin())"""
def init():
    game_result = battle()
    print(game_result)
    return "Game ended"

print(init())
```
